```python
import pandas as pd 
import pm4py as pm4 
from sklearn.metrics import mean_absolute_error 
from datetime import datetime 
import matplotlib.pyplot as plt
```

```python
df = pm4.convert_to_dataframe(pm4.read.read_xes('BPI_Challenge_2012.xes.gz'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df contains the DataFrame with columns 'concept:name' and 'time:timestamp'

# Convert 'time:timestamp' to datetime format
df['time:timestamp'] = pd.to_datetime(df['time:timestamp'], format = 'mixed')

# Filter data for January 2012
desired_month = 1  # January
df_january = df[df['time:timestamp'].dt.month == desired_month]

# Extract day of the week
df_january['day_of_week'] = df_january['time:timestamp'].dt.day_name()

# Reorder the days of the week
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_january['day_of_week'] = pd.Categorical(df_january['day_of_week'], categories=days_order, ordered=True)

# Group data by 'concept:name' and 'day_of_week', count occurrences
activity_day_counts = df_january.groupby(['concept:name', 'day_of_week']).size().unstack(fill_value=0)

# Calculate the average activity count per day of the week
average_activity_per_day = activity_day_counts.mean()

# Plot the histogram with day of the week on x-axis
plt.figure(figsize=(10, 6))
plt.bar(average_activity_per_day.index, average_activity_per_day, color='skyblue', alpha=0.7)
plt.xlabel('Day of the Week')
plt.ylabel('Average Activity Count')
plt.title('Average Activity Count per Day of the Week in January 2012')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('sunday_histogram.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df contains the DataFrame with columns 'concept:name', 'time:timestamp', and 'org:resource'

# Convert 'time:timestamp' to datetime format
df['time:timestamp'] = pd.to_datetime(df['time:timestamp'])

# Filter data for January 2012 Sundays
df_sundays_jan_2012 = df[(df['time:timestamp'].dt.year == 2012) & 
                          (df['time:timestamp'].dt.month == 1) & 
                          (df['time:timestamp'].dt.dayofweek == 6)]

# Filter data for 'org:resource' = 112 and count occurrences of each activity
df_resource_112 = df_sundays_jan_2012[df_sundays_jan_2012['org:resource'] == '112']
activity_counts_112 = df_resource_112['concept:name'].value_counts()

# Filter data for other 'org:resources' and count occurrences of each activity
df_other_resources = df_sundays_jan_2012[df_sundays_jan_2012['org:resource'] != '112']
activity_counts_other = df_other_resources['concept:name'].value_counts()

# Plot bar chart to compare activities done by org:resource 112 and other resources on Sundays
plt.figure(figsize=(15, 8))
activity_counts_112.plot(kind='bar', color='skyblue', position=0, width=0.4, label='Resource 112')
activity_counts_other.plot(kind='bar', color='orange', position=1, width=0.4, label='Other Resources')
plt.xlabel('Activity')
plt.ylabel('Count')
plt.title('Comparison of Activities Done on Sundays (January 2012)')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('activities_comparison_sundays_january_2012.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df contains the DataFrame with columns 'concept:name', 'time:timestamp', and 'org:resource'

# Convert 'time:timestamp' to datetime format
df['time:timestamp'] = pd.to_datetime(df['time:timestamp'])

# Extract day of the week
df['day_of_week'] = df['time:timestamp'].dt.day_name()

# Filter data for Sundays
df_sundays = df[df['day_of_week'] == 'Sunday']

# Group activities into categories: 'Application', 'Offer', 'Work'
# based on the starting letter of 'concept:name'
def categorize_activity(activity_name):
    if activity_name.startswith('A'):
        return 'Application'
    elif activity_name.startswith('O'):
        return 'Offer'
    elif activity_name.startswith('W'):
        return 'Work'
    else:
        return 'Other'

# Apply categorization function to create a new column 'activity_category'
df_sundays['activity_category'] = df_sundays['concept:name'].apply(categorize_activity)

# Count occurrences of each activity category
activity_category_counts = df_sundays['activity_category'].value_counts()

# Plot bar chart for activity categories done on Sundays
plt.figure(figsize=(10, 6))
activity_category_counts.plot(kind='bar', color='blue')
plt.xlabel('Activity Category')
plt.ylabel('Count')
plt.title('Activity Categories Done on Sundays')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.savefig('sundays_activity_categories_bar_chart.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df contains the DataFrame with columns 'concept:name', 'time:timestamp', and 'org:resource'

# Convert 'time:timestamp' to datetime format
df['time:timestamp'] = pd.to_datetime(df['time:timestamp'])

# Filter data for Sundays
df_sundays = df[df['time:timestamp'].dt.day_name() == 'Sunday']

# Filter data for 'org:resource' = 112 and count occurrences of each activity
df_resource_112 = df_sundays[df_sundays['org:resource'] == '112']
activity_counts_112 = df_resource_112['concept:name'].value_counts()

# Filter data for other 'org:resources' and count occurrences of each activity
df_other_resources = df_sundays[df_sundays['org:resource'] != '112']
activity_counts_other = df_other_resources['concept:name'].value_counts()

# Create a DataFrame to display numerical values
comparison_table = pd.DataFrame({
    'Resource 112': activity_counts_112,
    'Other Resources': activity_counts_other
})

# Plot the bar graph for comparison
plt.figure(figsize=(12, 8))
comparison_table.plot(kind='bar', color=['skyblue', 'orange'])  # Set color for each column
plt.xlabel('Activity')
plt.ylabel('Count')
plt.title('Comparison of Activities between Resource 112 and Other Resources on Sundays')
plt.xticks(rotation=90)
plt.yticks(range(0, 1300, 200))  # Set y-axis ticks from 0 to 1200 with a step of 200
plt.legend(title='Resource')
plt.tight_layout()
plt.savefig('comparison_112.png')
plt.show()
```
